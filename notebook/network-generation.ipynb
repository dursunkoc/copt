```python
import sys
import os
sys.path.insert(0, os.path.abspath('../src'))
import model_parameters as MP
import numpy as np
import networkx as nx
from networkx.algorithms.centrality import betweenness_centrality as BC
import matplotlib.pyplot as plt
from experiment.network_generator import gen_network

```

```python
def max_degree(grph):
    nn = max(grph.degree, key=lambda x: x[1])
    grph.remove_node(nn[0])
    return nn[0]

def max_degree_b(grph):
    betweeness = BC(grph)
    degrees = grph.degree
    measure = [(k,v,betweeness[k]) for (k,v) in degrees]
    nn = max(measure, key=lambda x: x[1]-x[2])
    grph.remove_node(nn[0])
    return nn[0]

def max_degree_c(grph):
    nn = max(grph.degree, key=lambda x: x[1])
    grph.remove_node(nn[0])
    return nn[0]

def next_for_max_degree(g):
    print(max_degree(g))
    plt.style.use('classic')
    nx.draw(g, pos=nx.fruchterman_reingold_layout(g), with_labels=True)
    plt.show()

def next_for_max_degree_b(g):
    print(max_degree_b(g))
    plt.style.use('classic')
    nx.draw(g, pos=nx.fruchterman_reingold_layout(g), with_labels=True)
    plt.show()

def next_for_max_degree_c(g):
    print(max_degree_c(g))
    plt.style.use('classic')
    nx.draw(g, pos=nx.fruchterman_reingold_layout(g), with_labels=True)
    plt.show()
```

```python
N_i_j, g = gen_network(n=20, seed=142, net_type='barabasi', m=3, p=None, drop_prob=.8)

print(np.sum(N_i_j))

plt.style.use('classic')
nx.draw(g, pos=nx.fruchterman_reingold_layout(g), with_labels=True)
plt.show()

for _ in range(len(g.nodes())):
    next_for_max_degree(g)
```
